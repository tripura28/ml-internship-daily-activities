File Handling functions

```python
f=open('abc.txt','w')
print('File name:',f.name)
print('mode:',f.mode)
print(f.readable())
print(f.writable())
print(f.closed)
f.close()
print(f.closed)
```

```python
with open("abc.txt","r") as file:
    a=file.read()
print(a)
```

```python
with open("abc.txt","r") as file:
    a=file.readline(5)
print(a)
```

```python
with open("abc.txt","r") as file:
    a=file.readlines()
print(a)
```

```python
with open("abc.txt","r") as file:
    a=file.read(4)
print(a)
```

```python
with open("abc.txt","w") as file:
    file.write("Aditya Degree College")
```

```python
l=["Aditya degree college\n","kakinada"]
with open("abc.txt","w")as file:
    file.writelines(l)
```
